### Bài 1. Phân Phối Nhị Thức (Binomial)

 Một bài thi trắc nghiệm bao gồm n câu, mỗi câu có m lựa chọn. Giả sử bạn không học bài và đánh lụi tất cả các câu hỏi. Có thể rằng giả định rằng xác suất lựa chọn các đáp án là như nhau và sự lựa chọn đáp án ở một câu hỏi không bị ảnh hưởng bởi các câu hỏi khác.

1.1 X là biến ngẫu nhiên thể hiện số câu đúng. X sẽ thuộc phân phối nào?

1.2 Tìm, dựng bảng phân phối xác suất và trực quan hóa (Bar chart) với các cặp (n,m) sau: (10,4), (50,4), (10,3), (50,3)

1.1 Biến ngẫu nhiên X thể hiện số câu đúng trong bài thi trắc nghiệm.
 Vì mỗi câu hỏi có m lựa chọn và bạn không học bài nên xác suất chọn đúng một câu hỏi là 1/m.
 Theo đó, X sẽ thuộc phân phối nhị thức (binomial distribution).

```python

 # 1.2
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Các cặp (n, m)
pairs = [(10, 4), (50, 4), (10, 3), (50, 3)]

# Vẽ biểu đồ cho từng cặp (n, m)
for n, m in pairs:
    # Tính phân phối xác suất
    x = np.arange(0, n + 1)
    p = 1 / m
    probability = binom.pmf(x, n, p)

    # Trực quan hóa bảng phân phối xác suất
    plt.figure()
    plt.bar(x, probability)
    plt.title(f'Probability Distribution for (n={n}, m={m})')
    plt.xlabel('Number of Correct Answers')
    plt.ylabel('Probability')
    plt.xticks(x)
    plt.show()
```

### Bài.2 Phân Phối Chuẩn (Normal)

#### Phát sinh mẫu

Giả sử chiều cao của nam thanh niên trưởng thành là biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 170cm, và độ lệch chuẩn là 4cm. Hãy vẽ phân phối hàm mật độ của phân phối trên. Tìm xác suất để một người trưởng thành có chiều cao nhỏ hơn 165 cm.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Tham số của phân phối chuẩn
mu = 170  # Kỳ vọng
sigma = 4  # Độ lệch chuẩn

# Tạo dữ liệu mẫu từ phân phối chuẩn
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100) # Tạo mảng x từ (mu - 3*sigma) đến (mu + 3*sigma)
pdf = norm.pdf(x, mu, sigma) # Tính hàm mật độ xác suất (PDF)

# Vẽ đồ thị hàm mật độ xác suất (PDF)
plt.plot(x, pdf, label='PDF')

# Tô màu diện tích dưới đường cong PDF từ x=mu-3*sigma đến x=165
plt.fill_between(x, pdf, where=(x < 165), color="red", alpha=0.5, label='P(X < 165)')

# Đặt tiêu đề và nhãn cho trục x và y
plt.title('Probability Density Function (PDF) of Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.grid(True)
plt.show()

# Tính xác suất P(X < 165) bằng cách tính diện tích dưới đường cong PDF từ x=mu-3*sigma đến x=165
probability = norm.cdf(165, mu, sigma)
print("Xác suất một người trưởng thành có chiều cao nhỏ hơn 165 cm là:", probability)
```
