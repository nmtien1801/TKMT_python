### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.
b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.
c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?
d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

import random
POP = [random.randint(120, 200) for _ in range(10000)]
print(POP)


#a
import numpy as np
POP_mean = np.mean(POP)
print("Chiều cao trung bình của quần thể là: ", POP_mean)
POP_std = np.std(POP)
print("Độ lệch chuẩn của quần thể là: ", POP_std)

#b
sample_greater_180 = [person for person in POP if person >=180] # count_tall = np.sum(POP >= 180)
print("Danh sách người cao (>180cm) trong quần thể:", sample_greater_180)
ratio_tall = len(sample_greater_180)/len(POP)
print("Tỉ lệ người cao trong quần thể (>180cm) là: ", ratio_tall*100,"%")

#c
import matplotlib.pyplot as plt
plt.hist(POP, bins=20)
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Số lượng người')
plt.title('Phân bố chiều cao của quần thể')
plt.show()

#d
from scipy.stats import norm
plt.hist(POP, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')

# Vẽ đồ thị hàm mật độ của phân phối chuẩn
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100) # Tạo mảng các giá trị cách đều nhau theo trục x
p = norm.pdf(x, POP_mean, POP_std) # Tính giá trị tại từng điểm của ppc
plt.plot(x, p)
plt.title('Phân phối chiều cao của quần thể và phân phối chuẩn')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Tần suất')
plt.legend(['Phân phối chuẩn', 'Quần thể'])
plt.show()
```

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu
a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
import numpy as np
import matplotlib.pyplot as plt


#a
# Số lần lấy mẫu
num_samples = 1000

# Kích thước mẫu
sample_size = 100

# Khảo sát tỷ lệ người cao
def simulate_height_proportion(sample_size):
    # Giả định phân phối chiều cao là một phân phối chuẩn (normal distribution) với trung bình 170 và độ lệch chuẩn 10
    heights = np.random.normal(170, 10, sample_size)
    # Đếm số người cao (chiều cao > 180)
    tall_count = np.sum(heights > 180)
    # Tính tỷ lệ người cao
    tall_proportion = tall_count / sample_size
    return tall_proportion

# Mô phỏng và tính toán tỷ lệ người cao cho các mẫu
proportions = [simulate_height_proportion(sample_size) for _ in range(num_samples)]

# Vẽ histogram
plt.figure(figsize=(8, 6))
plt.hist(proportions, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram của tỷ lệ người cao\n(Kích thước mẫu = 100)')
plt.xlabel('Tỷ lệ người cao(x100)')
plt.ylabel('Số lần xuất hiện')
plt.show()


#b
import numpy as np
import matplotlib.pyplot as plt

# Số lần lấy mẫu
num_samples = 1000

# Kích thước mẫu
sample_size = 500

# Khảo sát tỷ lệ người cao
def simulate_height_proportion(sample_size):
    # Giả định phân phối chiều cao là một phân phối chuẩn (normal distribution) với trung bình 170 và độ lệch chuẩn 10
    heights = np.random.normal(170, 10, sample_size)
    # Đếm số người cao (chiều cao > 180)
    tall_count = np.sum(heights > 180)
    # Tính tỷ lệ người cao
    tall_proportion = tall_count / sample_size
    return tall_proportion

# Mô phỏng và tính toán tỷ lệ người cao cho các mẫu
proportions = [simulate_height_proportion(sample_size) for _ in range(num_samples)]

# Vẽ histogram
plt.figure(figsize=(8, 6))
plt.hist(proportions, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram của tỷ lệ người cao\n(Kích thước mẫu = 100)')
plt.xlabel('Tỷ lệ người cao(x100)')
plt.ylabel('Số lần xuất hiện')
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
import numpy as np
import matplotlib.pyplot as plt

# Số lần lấy mẫu
num_samples = 1000

# Kích thước mẫu
sample_size = 100

# Khảo sát phương sai mẫu
def simulate_sample_variance(sample_size):
    # Giả định phân phối chiều cao là một phân phối chuẩn (normal distribution) với trung bình 170 và độ lệch chuẩn 10
    heights = np.random.normal(170, 10, sample_size)
    # Tính độ lệch chuẩn
    std_dev = np.std(heights)
    # Tính phương sai (bình phương độ lệch chuẩn)
    sample_variance = std_dev ** 2
    return sample_variance

# Mô phỏng và tính toán phương sai cho các mẫu
variances = [simulate_sample_variance(sample_size) for _ in range(num_samples)]

# Vẽ histogram
plt.figure(figsize=(8, 6))
plt.hist(variances, bins=20, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Histogram của phương sai mẫu\n(Kích thước mẫu = 100)')
plt.xlabel('Phương sai')
plt.ylabel('Số lần xuất hiện')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
from scipy.stats import t

# Dữ liệu
data = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]
n = len(data)  # Kích thước mẫu

# Trung bình mẫu
sample_mean = np.mean(data)

# Độ lệch chuẩn mẫu
sample_std = np.std(data, ddof=1)

# Độ tin cậy (95%)
confidence_level = 0.95

# a) Khi sigma đã biết
sigma = 0.3
z_critical = 1.96  # Z-score tương ứng với độ tin cậy 95%
margin_of_error_a = z_critical * (sigma / np.sqrt(n))
confidence_interval_a = (sample_mean - margin_of_error_a, sample_mean + margin_of_error_a)

# b) Khi sigma chưa biết
t_critical = t.ppf((1 + confidence_level) / 2, df=n - 1)  # T-score tương ứng với độ tin cậy 95% và n-1 bậc tự do
margin_of_error_b = t_critical * (sample_std / np.sqrt(n))
confidence_interval_b = (sample_mean - margin_of_error_b, sample_mean + margin_of_error_b)

print("a) Khoảng tin cậy cho trung bình (biết sigma):", confidence_interval_a)
print("b) Khoảng tin cậy cho trung bình (không biết sigma):", confidence_interval_b)
```
